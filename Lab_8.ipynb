1. Take two 3D coordinates from the user. Find out the distance between these points  using following distance functions.  


x and y are vectors representing the coordinates of two points.  
Here, k=3  


```python
def euclidean_distance(point1, point2):

  x1, y1, z1 = point1
  x2, y2, z2 = point2
  distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2 + (z2 - z1) ** 2) ** 0.5
  return distance

def manhattan_distance(point1, point2):

  x1, y1, z1 = point1
  x2, y2, z2 = point2
  distance = abs(x2 - x1) + abs(y2 - y1) + abs(z2 - z1)
  return distance

def minkowski_distance(point1, point2, p=3):

  x1, y1, z1 = point1
  x2, y2, z2 = point2
  distance = ((abs(x2 - x1) ** p) + (abs(y2 - y1) ** p) + (abs(z2 - z1) ** p)) ** (1/p)
  return distance

def get_point():

  while True:
    try:
      coordinates = input("Enter 3D coordinates (x, y, z) separated by spaces: ")
      x, y, z = map(float, coordinates.split())
      return (x, y, z)
    except ValueError:
      print("Invalid input. Please enter three numbers separated by spaces.")


point1 = get_point()
point2 = get_point()

euclidean_dist = euclidean_distance(point1, point2)
manhattan_dist = manhattan_distance(point1, point2)
minkowski_dist = minkowski_distance(point1, point2, p=2)  # Minkowski distance with p=2

print("Euclidean Distance:", euclidean_dist)
print("Manhattan Distance:", manhattan_dist)
print("Minkowski Distance (p=2):", minkowski_dist)
```

 2. Import the Iris dataset. Write a program to obtain the Euclidean Distance Matrix for  all the data samples in the feature space. Distance metric is a 2D array, where the  (i,j)th entry represents the distance between the ith and jth sample points in the feature  space.

```python
from sklearn.datasets import load_iris
from sklearn.metrics.pairwise import euclidean_distances

# Load the Iris dataset
iris = load_iris()
data = iris.data  # Features of the Iris dataset

# Calculate the Euclidean distance matrix
distance_matrix = euclidean_distances(data)

# Print the distance matrix shape
print("Distance matrix shape:", distance_matrix.shape)

# Print a sample sub-matrix (optional)
# You can adjust the slice range to view specific parts of the matrix
print("Sample sub-matrix:\n", distance_matrix[:5, :5])
```

3. Import the Iris dataset. Prepare a dataset considering samples belong to any two
output classes. Draw the scatter plot for all the samples in the new dataset considering
any two input attributes. Examine the scatter plot to find the equation of a line that
can separate sample of two classes.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target classes

# Select samples from two specific classes (e.g., class 0 and class 1)
class1 = 0
class2 = 1
X_new = X[(y == class1) | (y == class2)]  # Filter data for chosen classes
y_new = y[(y == class1) | (y == class2)]  # Filter target labels for chosen classes

# Extract two features for the scatter plot (e.g., features 0 and 1)
feature1 = 0
feature2 = 1
X_scatter = X_new[:, [feature1, feature2]]  # Select desired features

# Create the scatter plot
plt.scatter(X_scatter[:, 0], X_scatter[:, 1], c=y_new, cmap=plt.cm.coolwarm)  # Color by class
plt.xlabel(iris.feature_names[feature1])  # Label x-axis
plt.ylabel(iris.feature_names[feature2])  # Label y-axis
plt.title("Iris Dataset Scatter Plot (Classes {} and {})".format(class1, class2))
plt.show()
```
