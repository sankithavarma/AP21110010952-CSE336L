```python
#Q4 Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
```

```python
file="D:/sem-6/ML-Lab/week-2/datas.txt"
x='X-axis'
y='Y-axis'
title='Line plot of the values'
with open(file,'r') as f:
    a=f.readlines()
    b=[list(map(float,i.strip().split())) for i in a]
x_val=[item[0] for item in b]
y_val=[item[1] for item in b]
plt.plot(x_val,y_val,label='Line Plot',color='g')
plt.xlabel(x)
plt.ylabel(y)
plt.title(title)
plt.legend()
plt.show()
```

```python
#Q7 Create plot usingsubplots(),scatter, and some other methods on theAxesobject:
```

```python
x,y=np.random.rand(50),np.random.rand(50)
figure,axs=plt.subplots(2,2,figsize=(15,10))
axs[0,0].scatter(x,y,color='purple',label='Scatter Plot')
axs[0,1].plot(x,y,color='black',linestyle='--',marker='+',label='Line Plot')
bar_data={'Season 1':25,'Season 2':30,'Season':10}
axs[1,0].bar(bar_data.keys(),bar_data.values(),color='green',label='Bar Chart',edgecolor='black',width=0.5)
axs[1,1].hist(np.random.randn(1000),bins=25,color='red',label='Histogram',edgecolor='black')
plt.show()
```

```python
#Q8 Generate random data for x and y and create a scatter plot using Matplotlib.
```

```python
np.random.seed(30)
a=100
x=np.random.rand(a)
y=4*x+1+0.1*np.random.randn(a)
plt.scatter(x,y,label='Data',color='indigo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Random Data Points')
plt.legend()
plt.show()
```

```python
#Q11 Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
```

```python
categories = ['Rent', 'Utilities', 'Groceries', 'Entertainment', 'Transportation']
expenses = [1100, 300, 600, 200, 150]
plt.figure(figsize=(5, 5))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'gold', 'lightpink'])
plt.title('Expense Distribution in Budget')
plt.show()
```
